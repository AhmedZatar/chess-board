```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from random import random
```

```python
class ChessBoard:
    def __init__(self):
        self.grid=np.ones((8,8,3))
        self.red=[0,0]
        self.blue=[0,0]
        for i in range(len(self.grid)):
            for j in range(len(self.grid)):
                if i%2 == 0 and j%2==1:
                    self.grid[i,j]=[0,0,0]
                if i%2 == 1 and j%2 ==0:
                    self.grid[i,j]=[0,0,0]
    def add_red(self,row,column):
        self.grid[row,column]=[1,.2,0]
        self.red=[row,column]
    def add_blue(self,row,column):
        self.grid[row,column]=[0,1,1]
        self.blue=[row,column]
    def render(self):
        plt.imshow(self.grid)
        
    def is_under_attack(self):
        
        if self.blue[0]==self.red[0] or self.blue[1]==self.red[1]:
            return True
        if abs(self.blue[0]-self.red[0])==abs(self.blue[1]-self.red[1]):
            return True
               
        return False
        
        
        
    
    
        
        
```

```python
def test_same_row():
    chess = ChessBoard()
    chess.add_red(5,3)
    chess.add_blue(5,1)
    chess.render()
    assert chess.is_under_attack() == True
    print('Test pass')
    print('Under Attack!')

test_same_row()
    
```

```python
def test_same_column():
    chess = ChessBoard()
    chess.add_red(3,1)
    chess.add_blue(7,1)
    chess.render()
    assert chess.is_under_attack() == True
    print('Test pass')
    print('Under Attack!')

test_same_column()
```

```python
def test_same_diagonal():
    chess = ChessBoard()
    chess.add_red(5,3)
    chess.add_blue(3,1)
    chess.render()
    assert chess.is_under_attack() == True
    print('Test pass')
    print('Under Attack!')

test_same_diagonal()
```

```python
def test_not_under_arrack():
    chess = ChessBoard()
    chess.add_red(5,3)
    chess.add_blue(2,1)
    chess.render()
    assert chess.is_under_attack() == False
    print('Test pass')
    print('Not Under Attack!')

test_not_under_arrack()
```
